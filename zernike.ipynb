```python
wavelength=820*nm
size=20.0*mm
N=2000
print (size)
x = np.linspace(0,size,N)
print(len(x))
```

```python
F=Begin(size,wavelength,N)

```

```python
Noll = 6 # ordre du polynome 
(nz,mz) = noll_to_zern(Noll) # convert order to radial and azimuthal (nz,mz)=noll_to_zern(Noll)

A = 0.2
Z = Zernike(F,nz,mz,size/4,A,units ='lam')
(nz,mz) = noll_to_zern(7)
Z = Zernike(Z,nz,mz,size/4,0,units ='lam')
S = ZernikeName(Noll)
Z = CircAperture(size/4,0,0,Z)
Phi = Phase(Z)
plt.imshow(Phi, cmap='jet')
plt.title(S)
```

```python
F_FFT = PipFFT(Z)
I= Intensity(F_FFT)
plt.imshow(I, cmap='jet')
```

```python
L=Lens(Z,800*mm)
prog = Propagate (L,820*mm) #Forvard
I2 = Intensity(prog)
fig=plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.imshow(I2,cmap='jet')
ax1.set_xlim([900,1100])
ax1.set_ylim([900,1100])
print(len(I2))
ax2.plot(x*1E3,I2[1000])
ax2.set_xlabel('mm')
a.newDataReceived(I2)
```
